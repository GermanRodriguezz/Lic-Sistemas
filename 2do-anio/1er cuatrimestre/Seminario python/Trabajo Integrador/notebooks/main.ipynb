```python
import csv

mi_archivo = '../Archivos/usu_individual_T324.txt'
INDEX_CH04 = 11
INDEX_PONDERA = 9
INDEX_EDAD = 13
INDEX_CH10 = 17
INDEX_CH12 = 19
cantMujeres = 0
cantVarones = 0
SecundarioCompleto = 0
with open(mi_archivo, encoding= 'utf-8') as archivo_csv:
    reader = csv.reader(archivo_csv, delimiter=';')
    encabezado = next(reader)
    for row in reader :
        if row[INDEX_CH04] == '1' :
                cantVarones += int(row[INDEX_PONDERA]) 
        else :
            cantMujeres += int(row[INDEX_PONDERA])
        if row[INDEX_EDAD] > '18':
            if row[INDEX_CH10] == '2' :
                if row[INDEX_CH12] >= '4':
        
                    SecundarioCompleto += int(row[INDEX_PONDERA])
    print('Total de mujeres :' + str(cantMujeres) + '    Total de varones :'+ str(cantVarones))
    print('Total de personas con secundario completo :' + str(SecundarioCompleto))
```

```python
import csv
mi_archivo2 = '../Archivos/usu_hogar_T324.txt'
with open (mi_archivo2,encoding='utf-8') as archivo_csv2:
    reader = csv.reader(archivo_csv2,delimiter=';')
    encabezado = next(reader)
    INDEX_PONDERA = 8
    INDEX_VIVIENDA = 37
    INDEX_AGLOMERADO = 8
    INDEX_OCUPANTES = 64
    INDEX_BAÑO = 19
    viviendasTotales = 0
    viviendasQueCumplen = 0
    AglomeradoMax = -1 
    AglomeradoMin = 9999999
    for row in reader:
        if row[INDEX_VIVIENDA] == '1':
            viviendasQueCumplen += int(row[INDEX_PONDERA])
        else:
            viviendasTotales += int(row[INDEX_PONDERA])
        if row[INDEX_OCUPANTES] > '2' and row[INDEX_BAÑO] == '2':
            if int(row[INDEX_AGLOMERADO]) > int(AglomeradoMax) :
                AglomeradoMax = int(row[INDEX_AGLOMERADO])
            else :
                if int(row[INDEX_AGLOMERADO]) < int(AglomeradoMin) :
                    AglomeradoMin = int(row[INDEX_AGLOMERADO])
    print('El porcentaje de personas que viven en el terreno y propiedad a su nombre son ' + str((viviendasTotales/viviendasQueCumplen)*100))
    print('El aglomerado minino que cumple las condiciones es : ',AglomeradoMin)
    print('El aglomerado maximo que cumple las condiciones es : ',AglomeradoMax)
```
