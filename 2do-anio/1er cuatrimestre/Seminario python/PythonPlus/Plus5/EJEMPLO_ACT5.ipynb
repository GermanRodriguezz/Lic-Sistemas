```python
from pathlib import Path
from collections import Counter
from datetime import datetime
import csv
#recibe la columna
def enRango(data):
    formato = "%d/%m/%y, %H:%M:%S"
    # se almacena en h los datos de la columna Hora pero ahora con el formato
    h = datetime.strptime(data, formato)
    # devuelve las columnas que cumplen con la condicion
    return h.hour in range(8, 18)

#recibe el csv como un iterable,y si no recibe el 2do parametro, pone por defecto 1
def miFiltro(csv_reader, cant=1):
    try:
        # el paramtro de la funcion en Rango es una columna del csv
        # se guardaran las filas que su colummna Hora cumplan con la condicion
        users = filter(lambda x:  enRango(x["Hora"]), csv_reader)
        #retornara los CANT usuarios con mas ocurrencias en el iterable que se guarda en users
        return Counter( map(lambda x: x['Nombre completo del usuario'], users)).most_common(cant)
    except ValueError:
        print("El archivo CSV está vacío o no tiene el formato correcto.")
        return []
```

```python
file_route = Path('archivos') / 'logs_catedra.csv'
# recibe la ruta del archivo logs_catedra.csv
try: 
    #abre el archivo
    '''tira error al querer abrirlo por que archivo contiene caracteres no validos'''
    with open(file_route,encoding='utf-8') as file:
        #archivo en formato csv con diccionario
        csv_reader = csv.DictReader(file, delimiter=',')
        # x al finalizar la funcion tendra los dos usuarios con mas ocurrencias
        x = miFiltro(csv_reader, 2)
except FileNotFoundError:
    print(f'No se encontro el archivo {file_route}')
except UnicodeDecodeError:
    print(f'El archivo {file_route} contiene caracteres no validos')
except:
    print(f'Hubo un problema al abrir el archivo {file_route}')
else:
    #si no hubo errores, se ejecuta este else e imprime los resultados
    '''try agregado para evitar el error de que no se encuentren usuarios en el rango de horas especificado'''
    try:
        for elem in x:
            print(elem[0])
        print('AL FIN TERMINE')
    except IndexError:
        print('No se encontraron usuarios en el rango de horas especificado.')
```
