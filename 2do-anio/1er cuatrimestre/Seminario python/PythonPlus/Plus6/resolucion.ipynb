```python
"""
EJERCICIO 1. Escribí una función que devuelva todos los registros de los usuarios que
han realizado alguna actividad de teoría de python plus ordenados por Nombre
completo de usuario.
Nota: la información sobre las actividades de teoría de python plus se obtiene
verificando que la columna "Contexto del evento" contenga el texto 'ACT' y la columna
'Nombre evento' sea igual a 'Ha comenzado el intento'.
EJERCICIO 2. Utilizando la función anterior, escribí un programa que informe las
cantidades que realizaron las dos posibles actividades, identificando cada una con 'ACT
1' y ACT 2'."""

import pandas as pd

rutaArchivo = "archivos/logs_catedra.csv"

data = pd.read_csv(rutaArchivo)

index_contexto = "Contexto del evento"
index_nombre = "Nombre evento"


```

```python
# Filtrar los registros que contienen 'ACT' en el contexto y 'Ha comenzado el intento' en el nombre del evento
def filtrar_usuarios_actividad(data):
    # Filtrar los registros que cumplen con las condiciones aplicando mascara booleana
    data_filtrada = data[(data[index_contexto].str.contains('ACT')) & (data[index_nombre] == 'Ha comenzado el intento')]
    # Ordenar por Nombre completo de usuario
    data_ordenada = data_filtrada.sort_values(by='Nombre completo del usuario')
    
    return data_ordenada

data_filtrada = filtrar_usuarios_actividad(data)
```

# SOLUCION 2

```python
#print("Usuarios que han realizado actividades de teoría de Python Plus:")
#print(data_filtrada)  # Mostrar las primeras filas del DataFrame original para referencia

def contar_actividades(data_filtrada):
    # Contar las actividades
    # Diccionario para guardar las cantidades de cada actividad
    # Inicializar en 0 las actividades
    actividades = {
        'ACT 1': 0,
        'ACT 2': 0
    }
    # Pregunto el valor sobre la columna 'Contexto del evento' y voy acumulando las cantidades 
    for valor in data_filtrada['Contexto del evento']:
        if 'ACT 1' in valor:
            actividades['ACT 1'] += 1
        elif 'ACT 2' in valor:
            actividades['ACT 2'] += 1
    return actividades

resultados_actividades = contar_actividades(data_filtrada)


print("Cantidad de actividades realizadas:")
for actividad, cantidad in resultados_actividades.items():
    print(f"{actividad}: {cantidad}")
```
