**EJERCICIO 1.** Escribir una función que retorne toda la información de los 3 usuarios con más actividad nocturna del curso. Esta función recibe un parámetro opcional que indica el orden en que se retorna la información: "**A**", si se ordenan en forma ascendente por nombre de usuario, "**D**" si es descendente o **None** si no se aplica ningún orden en particular. Por defecto la función no aplica ningún orden. 

**IMPORTANTE:** consideramos que el horario nocturno es de 19:00 a 07:00.

**EJERCICIO 2.** Utilizando la función anterior, escribir un programa que muestre dos listados correspondiente a la actividad  de los 3 usuarios que más veces accedieron durante la noche: uno ordenado en forma ascendente por nombre de usuario y otro sin ningún orden específico.  

## Solución

### EJERCICIO 1

```python
import csv
from pathlib import Path
from datetime import datetime, time
import random

ruta_global = Path.cwd()
#print(ruta_global)
ruta_log = ruta_global / "Archivos" / "logs_catedra.csv"

def es_turno_noc (horario):
    return horario >= time(19,0) or horario < time(7,0)

def ordenar(archivo,orden = None):
    dicTurno = {}
    for fila in archivo:
        hora = fila["Hora"]
        try :
            hora_format = datetime.strptime(hora,"%d/%m/%y, %H:%M:%S")
            if es_turno_noc(hora_format.time()):
                usuario = fila["Nombre completo del usuario"]
                if usuario in dicTurno:
                    dicTurno[usuario] += 1
                else:
                    dicTurno[usuario] = 1
        except Exception as e:
            print(f"Error con la fila {fila}: {e}")
    if orden == "A":
        return sorted(dicTurno.items(),key = lambda x:x[1],reverse=True)[:3]
    elif orden == "D":
            return sorted(dicTurno.items(),key = lambda x:x[1],reverse=False)[:3]
    else:
        cantidad = min(3, len(dicTurno))
        return random.sample(list(dicTurno.items()),cantidad)


with open (ruta_log, encoding="utf-8") as archivo:
    cuerpo = csv.DictReader(archivo,delimiter=",")
    header = cuerpo.fieldnames
    orden = input("Ingrese cómo quiere el orden (A = mayor a menor, D = menor a mayor, Enter = aleatorio): ").strip()
    informacion = ordenar(cuerpo,orden)
    for usu,cant in informacion:
        print(f"{usu}: {cant} actividades")

        



```
